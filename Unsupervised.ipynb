```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
```

```python
cancer= load_breast_cancer()
# cancer
```

### Scaling

```python
X= cancer.data
```

```python
pd.DataFrame(X)
```

```python
scaler= StandardScaler()
scaler.fit(X)
X_scaled= scaler.transform(X)
```

### Information Extracted

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=30)
pca.fit(X_scaled)

X_pca=pca.transform(X_scaled)
```

```python
# pd.DataFrame(X_scaled)
```

```python
pd.DataFrame(X_pca)
```

```python
variance=[]

for i in range(0,30):
    variance.append(np.var(X_pca[:,i]))
```

```python
print(f'information extracted by 2 components: {np.sum(variance[:2])*100/np.sum(variance)} %')
```

```python
for i in variance:
    print(i/np.sum(variance)*100)
```

### Reducing Dimension

```python
pca=PCA(n_components=2)
pca.fit(X_scaled)
X_pca=pca.transform(X_scaled)
```

```python
print('original shape', X.shape, '\nreduced shape', X_pca.shape)
```

### Visualization of PCA result

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=X_pca[:,0], y= X_pca[:,1], hue= np.where(cancer.target,'benign','malignant'))

plt.xlabel('component 1')
plt.ylabel('component 2')
```

```python
* Information extracted is 63.24% 
* We reduced the dimension from 30 to 2 and lost 36.76% of the initiol information. Even so, the 2 components are enough to separate benign and malignant tumors
* this mean, wew can see the hidden pattern of the data
* malignant points is more spread that benign points
```

### Modelling with the help of PCA

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
#data
X=pd.DataFrame(X, columns=cancer.feature_names)
y= cancer.target
```

### Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    stratify=y,
    random_state=2020)
```

### Pipeline

```python
estimator=Pipeline([
    ('scaling', StandardScaler()),
    ('pca',PCA(n_components=2)),
    ('model',KNeighborsClassifier(n_neighbors=5))
])
```

### Model Fitting and Evaluation

```python
estimator.fit(X_train,y_train)
y_pred= estimator.predict(X_test)
print(classification_report(y_test,y_pred))
```

## Clustering K-Means (2)

```python
from scipy.spatial.distance import cdist, pdist
from sklearn.metrics import silhouette_score
```

```python
# Data
X= pd.read_csv('ilustrasi k means.csv', sep=';')
```

```python
X
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(
    x='tinggibadan',
    y='tinggibadanistri',
    data=X
)

plt.xlabel('tinggi badan')
plt.ylabel('tinggi badan istri')
```

berapa cluster yang terbaik?

### Determine the optimal number of cluster

#### Elbow Method

```python
# clustering variation
number_of_cluster= range(2,11)
clusterings= [KMeans(n_clusters=k).fit(X) for k in number_of_cluster]
centroids= [k.cluster_centers_ for k in clusterings]

# within sum square
D_k= [cdist(X,cent,'euclidean') for cent in centroids]
cIdx=[np.argmin(D,axis=1) for D in D_k]
dist=[np.min(D,axis=1) for D in D_k]
avg_withinSS=[sum(d)/X.shape[0] for d in dist]
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(number_of_cluster, avg_withinSS)
sns.scatterplot(number_of_cluster,avg_withinSS)

plt.xticks(number_of_cluster)

plt.xlabel('Number of clusters')
plt.ylabel('average within sum square')
```

from elbow method, it seems like the best n_clusters is 3

### silhouette score

```python
kmeans= KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
labels=kmeans.labels_
silhouette_score(X,labels,metric='euclidean')
```

```python
s_score=[]

number_of_cluster=range(2,11)

for i in number_of_cluster:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X)
    labels=kmeans.labels_
    s_score.append(silhouette_score(X,labels,metric='euclidean'))
```

```python
s_score
```

```python
plt.figure(figsize=(12,7))

sns.lineplot(number_of_cluster,s_score)
sns.scatterplot(number_of_cluster,s_score)

plt.xticks(number_of_cluster)
plt.xlabel('number of clusters')
plt.ylabel('sillhoutte score')
```

from silhouette score, the best number of clusters is also 3

### Final result

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
df=X.copy()
df['cluster']=kmeans.labels_
```

```python
plt.figure(figsize=(10,10))

sns.scatterplot(x='tinggibadan',y='tinggibadanistri',data=df,hue='cluster')
```

* cluster 0 adalah cluster yang berisi orang-orang dengan tinggi badan relatif pendek, dengan istri yang tinggi badannya juga relatif pendek.

* cluster 1 adalah cluster yang berisi orang-orang dengan tinggi badan cukup tinggi, dan memiliki istri dengan tinggi badan yang pendek-sedang

* cluster 3 berisikan orang-orang dengan tinggi badan tinggi dengan istri yang tinggi badannya juga tinggi.

## Agglomerative CLustering

```python
import pandas as pd
import numpy as np

from sklearn.cluster import AgglomerativeClustering

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import silhouette_score
```

### data

```python
df=pd.read_csv('ilustrasi k means.csv', sep=';')
```

```python
X=df.copy()
```

```python
X.head()
```

```python
plt.figure(figsize=(10,10))

sns.scatterplot(
    x='tinggibadan',
    y='tinggibadanistri',
    data=X
)

plt.xlabel('tinggi badan')
plt.ylabel('tinggi badan istri')

for i,txt in enumerate(range(0,df.shape[0])):
    plt.annotate(txt,(df['tinggibadan'][i]+0.3,df['tinggibadanistri'][i]+0.3), fontsize=12)
```

### Dendogram

```python
from scipy.cluster.hierarchy import dendrogram, ward, complete, average
```

```python
linkage_array=ward(X)

plt.figure(figsize=(10,8))

dendrogram(linkage_array,25)
ax= plt.gca()
bounds= ax.get_xbound()

ax.plot(bounds,[60,60],'--',c='k')
ax.plot(bounds,[30,30],'--',c='k')

ax.text(bounds[1],60,'two clusters',va='center', fontdict={'size':15})
ax.text(bounds[1],30,'three clusters',va='center', fontdict={'size':15})

plt.title('Ward')
plt.xlabel('Sample Index')
plt.ylabel('cluster distance')

plt.show()
```

```python
linkage_array=average(X)

plt.figure(figsize=(10,8))

dendrogram(linkage_array,25)
ax= plt.gca()
bounds= ax.get_xbound()

ax.plot(bounds,[25,25],'--',c='k')
ax.plot(bounds,[15,15],'--',c='k')

ax.text(bounds[1],25,'two clusters',va='center', fontdict={'size':15})
ax.text(bounds[1],15,'three clusters',va='center', fontdict={'size':15})

plt.title('Average')
plt.xlabel('Sample Index')
plt.ylabel('cluster distance')

plt.show()
```

```python
linkage_array=complete(X)

plt.figure(figsize=(10,8))

dendrogram(linkage_array,25)
ax= plt.gca()
bounds= ax.get_xbound()

ax.plot(bounds,[40,40],'--',c='k')
ax.plot(bounds,[20,20],'--',c='k')

ax.text(bounds[1],40,'two clusters',va='center', fontdict={'size':15})
ax.text(bounds[1],20,'three clusters',va='center', fontdict={'size':15})

plt.title('Complete')
plt.xlabel('Sample Index')
plt.ylabel('cluster distance')

plt.show()
```

Results Comparison

```python
agg_ward=AgglomerativeClustering(n_clusters=3,linkage='ward')
df['ward']=agg_ward.fit_predict(X)

agg_ward=AgglomerativeClustering(n_clusters=3,linkage='average')
df['average']=agg_ward.fit_predict(X)

agg_ward=AgglomerativeClustering(n_clusters=3,linkage='complete')
df['complete']=agg_ward.fit_predict(X)
```

```python
df
```

```python
plt.figure(figsize=(10,10))

sns.scatterplot(x='tinggibadan',y='tinggibadanistri', data=df, hue='ward')

plt.xlabel('tinggi badan')
plt.ylabel('tinggi badan istri')
```

## DBSCAN

```python
df=pd.read_csv("ilustrasi k means.csv",sep=';')
```

```python
X=df.copy()
```

```python
plt.figure(figsize=(10,10))

sns.scatterplot(
    x='tinggibadan',
    y='tinggibadanistri',
    data=X
)

plt.xlabel('tinggi badan')
plt.ylabel('tinggi badan istri')
```

### Initial CLustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
dbscan= DBSCAN(eps=0.8, min_samples=6)
cluster=dbscan.fit_predict(X_scaled)
```

```python
print(cluster)
```

```python
silhouette_score(X_scaled,cluster)
```

Optimizing minimum samples and epsilon

```python
for eps in [i/10 for i in range(6,8)]:
    for min_samples in range (4,6):
        print(f'\neps {eps}')
        print(f'min samples {min_samples}')

        dbscan=DBSCAN(eps=eps, min_samples=min_samples)
        labels= dbscan.fit_predict(X_scaled)
        score= silhouette_score(X_scaled,labels)

        print(f'clusters present: {np.unique(labels)}')
        print(f'Cluster sizes: {np.bincount(labels +1)}')
        print(f'Silhouette Score : {score}')
```

```python
the best hyperparameters are eps: 0.7 and min samples=4, because no samples is included as noise (easier to inteprete) and has highest silhouette score.
```

### Final result

```python
dbscan=DBSCAN(eps=0.7, min_samples=4)
labels= dbscan.fit_predict(X_scaled)
```

```python
df['cluster']=labels
```

```python
plt.figure(figsize=(10,10))

sns.scatterplot(
    x='tinggibadan',
    y='tinggibadanistri',
    data=df, hue='cluster'
)

plt.xlabel('tinggi badan')
plt.ylabel('tinggi badan istri')
```
